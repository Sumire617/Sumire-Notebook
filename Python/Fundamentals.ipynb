```python
# 函数

# 默认参数n=2
def Default(n=2):
    total = 0
    for _ in range(n):
        total += _
    return total


# 可变参数
def add(*args):
    total = 0
    for arg in args:
        total += arg
    return total


print(Default())
print(Default(5))
print(add(1, 2, 3))
```

```python
# 类


class Class:
    # __slots__ 是一个特殊的类属性，用于限制类的实例可以拥有的属性。这里将实例属性限制为 _age、_name 和 _sex，可以节省内存并提高访问速度。
    __slots__ = {"_age", "_name", "_sex"}

    # 初始化方法
    def __init__(self, name=None, age=None):
        # _表示保护属性
        self._name = name
        self._age = age

    # @property 是Python的装饰器，用于将方法转换为只读属性。通过访问器，可以以属性的方式访问类的私有属性，而不需要调用方法。
    # 访问器
    @property
    def age(self):
        return self._age

    @property
    def name(self):
        return self._name

    @age.setter
    def age(self, age):
        self._age = age

    @name.setter
    def name(self, name):
        self._name = name

    # 定义对象的可读字符串表示,用于print()或者str()
    def __str__(self):
        return f"name = {self._name}, age = {self._age}"

    # 定义对象的“开发者友好字符串”表示,用于调试和 repr()
    def __repr__(self):
        return f"name = {self._name}, age = {self._age}"

    # 允许对象像函数一样被调用
    def __call__(self, name):
        return f"name = {self._name}, age = {self._age}"

    # 允许下标访问
    def __getitem__(self, index):
        return self._name[index]

    # 允许赋值
    def __setitem__(self, index, value):
        self._name[index] = value

    # 允许删除
    def __delitem__(self, index):
        del self._name[index]

    # @staticmethod 是Python的装饰器，用于定义静态方法。静态方法不需要访问类的实例或类本身，因此不需要 self 或 cls 参数。
    # 静态方法通常用于执行与类相关的独立操作，例如验证输入的性别是否合法。
    @staticmethod
    def is_true(sex):
        if sex != "男" and sex != "女":
            return False

    # @classmethod 是Python的装饰器，用于定义类方法。类方法需要访问类本身，因此需要 cls 参数。

    """
    # __add__：重载 + 运算符。
    # __sub__：重载 - 运算符。
    # __mul__：重载 * 运算符。
    # __eq__：重载 == 比较操作。
    # __lt__：重载 < 比较操作。
    # __len__：定义 len() 的行为。
    # __iter__：使对象可迭代。
    # __next__：定义迭代器的行为。
    # __enter__ 和 __exit__：实现上下文管理（with 语句）
    # __del__：析构方法，在对象被垃圾回收时调用。
    """


class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    @property
    def name(self):
        return self.__name

    @property
    def age(self):
        return self.__age

    def __str__(self):
        return f"name = {self.__name}, age = {self.__age}"

    @name.setter
    def name(self, name):
        self.__name = name

    @age.setter
    def age(self, age):
        self.__age = age


p = Person("Sumire", 18)
# print(p.__name)无法访问私有属性
# 可以通过print(p._Person__name)访问私有属性,但不建议这样做,因为这是Python的一种约定,并不是强制的
# 另一种推荐做法是使用@property装饰器,将私有属性封装成只读属性,这样可以防止外部直接访问和修改私有属性
print(p.name)
# 要修改私有属性,可以使用setter装饰器
p.name = "BUG"
print(p.name)
```

```python
# 迭代器与生成器
class Iterator:
    list = [1, 2, 3, 4]
    it = iter(list)
    # 可使用next()函数返回迭代器的下一个元素,迭代器下一个元素为1
    first = next(it)

    def fibonaci(self, n):
        a, b, counter = 0, 1, 0
        while True:
            if counter > n:
                return
            # 使用了yield的函数即生成器
            """
            yield: 不会退出的return,会返回一个生成器对象,可以使用for循环遍历
            每次遍历都会执行一次函数,直到遇到yield关键字,然后返回yield后面的值
            """
            yield a
            a, b = b, a + b
            counter += 1
```

```python
# 文件操作
class File:
    def __init__(self, file_path):
        self.file_path = file_path

    def read_file(self):
        """
        以文本模式读取文件内容
        :return: 文件内容
        """
        try:
            with open(self.file_path, "r", encoding="utf-8") as file:
                return file.read()
        except FileNotFoundError:
            print(f"文件 {self.file_path} 未找到。")
        except Exception as e:
            print(f"读取文件时发生错误: {e}")

    def write_file(self, content):
        """
        以文本模式写入内容到文件
        :param content: 要写入的内容
        """
        try:
            with open(self.file_path, "w", encoding="utf-8") as file:
                file.write(content)
            print(f"内容已成功写入 {self.file_path}。")
        except Exception as e:
            print(f"写入文件时发生错误: {e}")

    def append_file(self, content):
        """
        以文本模式追加内容到文件
        :param content: 要追加的内容
        """
        try:
            with open(self.file_path, "a", encoding="utf-8") as file:
                file.write(content)
            print(f"内容已成功追加到 {self.file_path}。")
        except Exception as e:
            print(f"追加文件时发生错误: {e}")

    def read_binary_file(self):
        """
        以二进制模式读取文件内容
        :return: 文件内容
        """
        try:
            with open(self.file_path, "rb") as file:
                return file.read()
        except FileNotFoundError:
            print(f"文件 {self.file_path} 未找到。")
        except Exception as e:
            print(f"读取二进制文件时发生错误: {e}")

    def write_binary_file(self, content):
        """
        以二进制模式写入内容到文件
        :param content: 要写入的二进制内容
        """
        try:
            with open(self.file_path, "wb") as file:
                file.write(content)
            print(f"二进制内容已成功写入 {self.file_path}。")
        except Exception as e:
            print(f"写入二进制文件时发生错误: {e}")
```

```python
# 网络编程
import requests
from threading import Thread


# ()里的内容指的是一个类的父类,即这个类继承的其他类
# 如果不需要继承其他类,可以将()留空
class Download(Thread):
    def __init__(self, url):
        # 传入url参数
        self.url = url
        # super()表示获取当前类的父类,并调用它的构造函数
        # 即继承其父类的构造函数并添加一个额外参数url
        super().__init__()

    def run(self):
        # self.url.rfind('/'):找到url最后一个/出现的位置(指针)
        # self.url.rding('/'+1):找到后+1,将指针移动到/后,这样就可以获取文件名,避免获取到/字符
        filename = self.url[self.url.rfind("/") + 1 :]
        # 向给定的self.url发送HTTP GET请求,将响应存储在resp的变量中
        # 这个操作通常用来获取网页或者从Web API中获取数据。
        # 请求可以包括头信息,查询参数,请求正文等
        # 通过使用resp.text(Unicode文本数据)、resp.content(binary)等方法可以获取响应的文本内容
        resp = requests.get(self.url)
        # 在二进制模式下打开文件
        # 'wb':即write和binary,表示写入二进制数据
        # 使用with语句可以在代码执行结束后自动关闭打开的文件
        with open("/User/16673/" + filename, "wb") as f:
            # resp.content返回的是二进制数据,二进制文件,比如音频,视频,图片等文件
            # 如果内容为文本内容,用resp.text更合适
            # 如果响应内容非非文本内容,用resp.content更合适
            f.write(resp.content)


def main():
    params = {
        "key": "31a4fa04bcd447894c20f5fa72114e64",
        "url": "https://www.tianapi.com/article/62",
    }
    resp = requests.get("https://apis.tianapi.com/htmlpic/index", params=params)
    date_model = resp.json()
    for mm_dict in date_model["result"]["list"]:
        url = mm_dict["picUrl"]
        print(url)


if __name__ == "__main__":
    main()
```
